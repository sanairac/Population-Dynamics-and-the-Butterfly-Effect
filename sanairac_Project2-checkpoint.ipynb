# The Mayfly Model: Population Dynamics and the Butterfly Effect

Physical systems are characterised by seemingly erratic tendencies, which gives rise to the study of chaotic systems. Chaotic systems are always deterministic and may be very simple, but produce divergent behaviour due to hypersensitivity to changes in initial conditions. The "butterfly effect" describes systems where a small change in the initial conditions can result in large differences in outcome at a later time. These characterizations can be visualized when attempting to predict changes in the population of people, animals, insects, etc. For the purposes of this report we will be using the "Mayfly Model" as an example, constructing a series of simple population models, examining the convergence or divergence of trajectories. 

Mayflies are aquatic organisms which spend most of the year at the bottom of a water body, emerging to the surface for a few days each year in order to mate, lay eggs and die. In effect, only one generation of mayflies is alive at anytime. We will assume that size of the population in a given year depends on the size of the population the previous year. An exponential model of growth or decay can be constructed assuming that $g$ is a constant function $g(x) = a$, for some fixed constant $a$, which results in the formula $x_n = ax_{n-1}$. Inductively, $x_n = a^nx_0$ can be dervied from this formula. However, exponential models cannot be effective for long-term predictions, since the population will effectively spiral to infinity or will decay to zero and this is not an effective representation of actual populations which compete for food and resources. Thus, when tracking the dynamics of population change, the "mayfly model" will be constructed using the following generalized equation:

$$ x_{t+1} = b(1-x_t)x_t $$
  
where $x_t$ is the proportion of mayfly population in year $t$ to the maximum possible population, while $b$ stands for the maximum growth rate. 

We will use Python to visualise multiplicative effects between componensts, fractals and predictive models, while studying deterministic dynamical systems. These are systems which develop over time, where the current state completely determines the future. The mayfly model supposedly follows a deterministic model of growth. By being in control of the parameters, which are $x$, $b$ and $t$, we will be able to explore the characteristics of the "mayfly model" and construct our own observations in order to ultimately be able to explain the "butterfly effect". Our methods of observation will include plotting population dynamics over time, with a fixed maximum growth rate($b$) and a fixed initial population($x$), respectively. In order to uncover interesting behaviors, we might repeat certain tasks while using different values.

We will begin by importing the modules and packages that will define and implement the functions, classes, or variables that will be used over the course of this report:

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
from matplotlib import pylab
%matplotlib inline
```

## For fixed values of b, x varies against t

Our first approach to exploring the "mayfly model" will be to study the dynamic effects of a fixed growth rate on a population over a period of twenty years.  

```python
def logistic(b,x):                       # the logistic function will be used repeatedly to run our generalized formula
    return b*x*(1 - x)       # x is the initial value of the population, b is the growth factor, t is the number of years  

def mayfly(x,b,t):         #creating the mayfly function so we can quickly and effectively produce plots using different values
    plt.plot(0,x, 'ro', clip_on=False)   #accounting for the initial value of x on the plot
    for i in range(t):
        x=logistic(b,x)
        plt.plot(i+1,x, 'ro',clip_on=False)
    plt.title("Population proportion vs Time")
    plt.ylim(0,1)                        #setting limits for the x and y axes so the plots can be easily compared
    plt.xlim(0,20)
    plt.xlabel("Year")
    plt.ylabel("Population proportion")
    plt.show()

mayfly(0.45,1.25,20)  #Test 1
```

Test $1$ illustrates a decay in the population over the course of $20$ years with a fixed maximum growth rate of $1.25$ and an initial population proportion of $0.45$. As we see from the graph, the mayfly population declines over time and gradually converges to an asymptote.

To see if the same pattern holds true we plot as follows in Test 2, entering another $b$ value while keeping the same value for $x$ and $t$.

```python
mayfly(0.45,2.2,20)  #Test 2
```

In Test $2$ we raised the growth rate to $2.2$ and there is a sudden growth in the population after which it seems to remain static over the period of $20$ years. 

```python
mayfly(0.45,2.9,20)  #Test 3
```

In Test $3$ we begin to observe a cycle of population growth and decay, before the population dynamics gradually stabilize in the latter half of the $20$ year period. 

We again enter a different value for $b$ in Test $4$.

```python
mayfly(0.45,3.2,20) #Test 4
```

With the growth rate at $3.2$ in Test $4$, a "boom-bust cycle" appears to repeat consistently over the course of the $20$ years we are tracking, with the population growing and declining rapidly with each generation.

```python
mayfly(0.45,3.8,20) #Test 5
```

Test $5$ is particularly interesting because it appears that for a high growth rate of $3.8$ there is no generalizable pattern. The population dynamics are characterized by irregular growth and decline with each successive generation.

After observing a pattern of behavior for the different values of $b$ at the fixed initial population of $x$, we will now observe whether a slight difference in the values of the initial population($x$) can result in variations in the pattern. I have selected the $b$ value of $3.2$ used in Test $4$, in order to observe whether the "boom-bust" cycle will be repeated.

```python
b = 3.2                          #using the same b value from Test 4
x1 = 0.35                        #using different a different x values from the one used in Test 4 
x2 = 0.25                        #using another x value which is only slightly different from the first
plt.plot(0,x1,'ro', ms=8)        #plotting the starting points of x & y
plt.plot(0,x2,'bo', ms=5)
for t in range(40):      #plotting over the course of 40 years in order to decisively confirm the presence of a pattern if any 
    x1 = b*(1-x1)*x1
    x2 = b*(1-x2)*x2
    plt.plot(t+1,x1,'ro', clip_on=False, ms=8)  #using different marker sizes and colors to differentiate between x1 & x2
    plt.plot(t+1,x2,'bo', clip_on=False, ms=5)
plt.title("Population proportion vs Time")
plt.xlabel("Year")
plt.ylabel("Population proportion")
plt.show()
```

The above graph shows that regardless of where the two initial population values started($x1$ and $x2$), at a growth rate of $3.2$ both $x1$ and $x2$ entered the pattern of a "boom-bust" cycle previously observed in Test $4$, indicating that this pattern can be predicted to occur at a certain growth rate.

We will now explore a new system of behaviors, where the growth rate shall remain fixed, while the population varies.

## For fixed values of x, b varies against t

```python
def bplot(m,n,x):   #m and n are the paremeters of the growth rate b, while x is the population proportion
    tmax=500        
    b=linspace(m,n,100)
    for i in range(tmax):
        if i>250:
            plt.plot(b,x, 'g.', alpha=.15, ms=2)
        x=b*(1-x)*x
    plt.xlabel("b value")
    plt.ylabel("x")
    plt.title("x versus b")
    plt.show()
    
bplot(1,3,0.4)  #Test
```

Here we plot the stabilized $x$ over a $100$ $b$ values varying from $1$ to $3$. As we can see, each $b$ value corresponds to a distinct, unique $x$ value. It means that for each $b$ value in $(1,3)$, the population dynamics in the mayfly model will eventually settle down to an asymptotic state, converging to a unique $x$ value. We do not see a "boom-bust" here. This observation matches what we have seen in the first half of this report, where all plots eventually show a stabilized, converging $x$.

Now we proceed to test different $b$ values in a different range to see if we can find something more interesting in a bifurcation diagram.

```python
b = np.linspace(1,4.0,10000)        # b acts as parameter and x as the initial value
x = 0.5*np.ones(10000)
    
plt.figure(figsize=(10,5))
for i in range(1000):
    x = logistic(b,x)
    if i >= 900:
        plt.plot(b,x, ',k',clip_on=False, alpha=0.02, ms=0.7)
    plt.xlim(1,4)                   #setting a limit for the x axes in order to producer a clearer graph

plt.xlabel('Reproduction Rate', fontsize= 10)
plt.ylabel('Population', fontsize= 10)
plt.title('Bifurcation Diagram', fontsize=20)
plt.show()
plt.show()
```

By plotting the stabilized $x$ values for a $100$ $b$ values in from $0$ to $4$ we begin to see more regularity for $b$ values between $1$ and $3$, but the stabilized $x$ behaves quite unexpectedly. 
Between $3$ and $3.4$ the $x$ is stabilized to two points and between $3.4$ and $3.5$ it stabilizes to $4$ points.

Note all the range boundaries here are just approximate values. We will now zoom in to observe what is happening:

```python
plt.figure(figsize=(10,5))
def phases (b,x):           #this function will repeatedly produce the bifurcation diagram for us at different ranges of values
    for t in range(500):
        x=logistic(b,x)
        plt.plot(b,x,',r',clip_on=True,markersize=2, alpha=0.15)
    plt.xlabel('Reproduction Rate', fontsize= 10)
    plt.ylabel('Population', fontsize= 10)
    plt.title('Mayfly Model', fontsize=20)
    plt.show()

b=linspace(3,4,700)
x = 0.5*ones(700) 
phases (b,x)
```

Based on this graph, there appears to a be number of discrete paths the population can potentially take. There is clearly a pattern of branches occuring within this population model, where the population successively takes on two discrete paths at each branch. It is still hard to say $x$ stabilizes for a continuity of values before we zoom in closer. 

```python
plt.figure(figsize=(10,5))
b=linspace(3.43,3.6,700)
x = 0.5*ones(700)
plt.ylim(0.8,0.9)
phases(b,x)
```

It is clear that for specific values of $b$, such as $3.4$ and $3.54$(these are estimates), we must expect a population "boom-bust" cycle where the system essentially oscillates exclusively between two population values, until the population branches into two discrete paths.

```python
plt.figure(figsize=(10,5))
b=linspace(3.55,3.6,700)
x = 0.5*ones(700)
plt.ylim(0.3,0.6)
phases(b,x)
```

After further zooming in we can deduce that this system is effectively able to land on almost any population value. The oscillations continue to occur in consistently smaller periods. This is a fractal system of chaos, which follows the deterministic rule- by knowing a certain generation's population value, we can determine the next.

The "butterfly effect" has come into play in these population dynamics, as we have observed the points of sudden divergence or convergence comes in our graphs. These systems have been sensitive to their initial conditions and we shall study this through one last model:

```python
x= 0.30000001        #miniscule differences in our initial values
y= 0.3
b=4

plt.figure(figsize=(8,8))
ax1= plt.subplot(2,1,1)
ax2= plt.subplot(2,1,2)
ax1.plot(0,y,'bo')
ax1.plot(0,x,'ro')     
for t in range(40):
        x= b*(1-x)*x
        y= b*(1-y)*y
        ax1.plot(t+1,x,'ro', clip_on=False, ms=6)
        ax1.plot(t+1,y,'bo', clip_on=False, ms= 4)
        ax2.semilogy(t+1,(abs(y-x)), 'ko', ms=4)

ax1.set_xlabel('Time')
ax2.set_ylabel('Time')
ax1.set_ylabel('trajectories x & y')
ax2.set_ylabel('separation between trajectories');
```

Subplot $1$ shows us that there were no differences between the trajectories of $x$ and $y$, but after the year $20$ these trajectories diverge and the difference between them rapidly increases as shown in subplot $2$. This is reflective of the "butterfly effect".
